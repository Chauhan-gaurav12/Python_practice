```python
# local and Global Variable 
# local variable are restricted to only one block of code and cannot
#     be changed throughout the program

# Global variable - global variable are not restricted to one cpde of 
#       block of code and can be chaned throughout the program 
x=145
print(x)

def loca():
    global x
    x=43
    return x
print(loca())
print(x)
```

```python
def add_numbers(numbers):
    total = 0
    for num in numbers:
        total = total + num
    print("The sum of the list is:", total)

my_list = []
for i in range(5):
    num = int(input(f"Enter number {i+1}: "))
    my_list.append(num)

add_numbers(my_list)
```

```python
# using recursion

def rec(numbers):
    if len(numbers)==1:
        return (numbers[0])
    else:
        return (numbers[0])+rec(numbers[1:])

print(rec([1,2,3,4,5]))
```

```python
def fib(num):
    if num== 1:
        return(0)
    elif num==2:
        return 1
    else:
        return(fib(num-1)+fib(num-2))
num=int(input("Enter the number: "))
fib(num)
```
